```python
#imports libraries
import pandas as pd
import requests
pd.set_option('display.max_columns', None)

```

```python
#writing an html document with the html from the website
url = "https://www.basketball-reference.com/leagues/NBA_2023_totals.html"
r = requests.get(url = url)
with open("bball_file/2022.html", "w+") as f:
    f.write(r.text)
    

```

```python
#import beautiful soup 
from bs4 import BeautifulSoup

```

```python
#opening the html file written from before.
with open("bball_file/2022.html") as f:
    page = f.read()
```

```python
#use beautiful soup to parse through the page file 
soup = BeautifulSoup(page, "html.parser")
```

```python
#looks for all rows with the thead class and removes them. 
theads = soup.find_all('tr', class_='thead')
for thead in theads:
    thead.decompose()
```

```python
#locating the table 
stats = soup.find(id = "totals_stats")
```

```python
#import pandas
import pandas as pd
```

```python
#creating a dataframe using pandas and the table we located. 
total_stats = pd.read_html(str(stats))[0]
df = pd.DataFrame(total_stats)
df

```

```python
#converting the datatypes of the columns of the dataframe to numerical datatypes.
df = df.apply(pd.to_numeric, errors = "ignore")
```

```python
#checking to see if any error rows are still present. 
df["Player"].unique()
```

```python
#convert the dataframe to a csv file. 
df.to_csv("nba_stats.csv")
```

```python
#deleting a column that will not be used. it came from the imported csv file. 
del players["Unnamed: 0"]
```

```python
#data cleaning for players that are on multiple teams (i.e. they have multiple rows). 
def get_single(data): 
    if data.shape[0] == 1: #if one row, return the data frame
        return data 
    else:
        row = data[data["Tm"] == "TOT"] #to get the total row only 
        row["Tm"] = df.iloc[-1,:]["Tm"] #rename total row team to current team name
        return row

players = players.groupby(["Player"]).apply(get_single) #applying the function to all rows
players
```

```python
players.index = players.index.droplevel() #drops a level of index from the table
```
